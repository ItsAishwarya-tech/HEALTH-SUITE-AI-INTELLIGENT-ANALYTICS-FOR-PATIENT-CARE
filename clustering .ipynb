```python
 # Define columns to keep
# # ---------------------------


patient_cols = ['patientunitstayid', 'gender', 'age', 'ethnicity',
        'admissionheight', 'admissionweight', 'dischargeweight',
        'apacheadmissiondx', 'hospitaladmitsource',
        'hospitaldischargestatus', 'unitstaytype']

diagnosis_cols = ['patientunitstayid', 'diagnosisstring','icd9code',
                  'diagnosispriority']

# # ...existing code...
apache_cols = [
    'patientunitstayid', 'acutephysiologyscore', 'apachescore',
    'predictedicumortality', 'predictediculos', 'predictedhospitalmortality',
    'predictedhospitallos'
]

apache_var_cols = [
    'patientunitstayid', 'sicuday', 
    'saps3day1', 'saps3today', 'saps3yesterday', 
    'gender', 'admitsource', 'age', 
    'admitdiagnosis', 'thrombolytics',
    'aids', 'hepaticfailure', 'lymphoma', 'metastaticcancer', 'leukemia',
    'immunosuppression', 'cirrhosis', 'electivesurgery', 'diabetes',
    'ventday1', 'oobventday1', 'oobintubday1',
    'pao2', 'fio2', 'day1pao2', 'day1fio2',
    'ejectfx', 'creatinine',
    'meds', 'verbal', 'motor', 'eyes', 
    'day1meds', 'day1verbal', 'day1motor', 'day1eyes'
]

vitals_cols = [
    'patientunitstayid',
    'temperature', 'sao2', 'heartrate', 'respiration',
    'systemicsystolic', 'systemicdiastolic', 'systemicmean'
]
medication_cols = [

    'patientunitstayid', 'drugname', 'dosage', 
    'routeadmin', 'frequency', 'prn'
]

custom_lab_cols = [
'patientunitstayid', 'labothername', 'labotherresult'
]
nurse_assess_cols = [
'patientunitstayid','celllabel','cellattributevalue'
]

past_history_cols = [
    'patientunitstayid', 'pasthistoryvalue'
]
infusion_cols = [
    'patientunitstayid', 'drugname', 'drugrate',
    'infusionrate', 'drugamount', 'volumeoffluid'
]


```

```python
# Apply filters
# ---------------------------
patient = patient[patient_cols]
diagnosis = diagnosis[diagnosis_cols]
apache = apache[apache_cols]
apache_var = apache_var[apache_var_cols]
vitals_cols = vital_pred[vitals_cols]
medication = medication[medication_cols]
custom_lab = custom_lab[custom_lab_cols]
nurse_assess = nurse_assess[nurse_assess_cols]
past_history = past_history[past_history_cols]
infusion_drug = infusion_drug[infusion_cols]

```

```python
# Summarize counts [ for feature engineering ]
diagnosis_summary = diagnosis.groupby('patientunitstayid').size().reset_index(name='diagnosis_count')
apache_summary = apache.groupby('patientunitstayid').size().reset_index(name='apache_count')
nurse_assess_summary = nurse_assess.groupby('patientunitstayid').size().reset_index(name='nurse_assess_count')
apache_var_summary = apache_var.groupby('patientunitstayid').size().reset_index(name='apache_var_count')
vitals_cols_summary = vital_pred .groupby('patientunitstayid').size().reset_index(name='vitals_count')
medication_cols_sumary = medication.groupby('patientunitstayid').size().reset_index(name='medication_count')
custom_lab_summary = custom_lab.groupby('patientunitstayid').size().reset_index(name='custom_lab_count')
past_history_summary = past_history.groupby('patientunitstayid').size().reset_index(name='past_history_count')
infusion_drug_summary = infusion_drug.groupby('patientunitstayid').size().reset_index(name='infusion_drug_count')


```

```python
# ...existing code...
merged = patient.merge(apache_summary, on='patientunitstayid', how='left')
merged = merged.merge(diagnosis_summary, on='patientunitstayid', how='left')
merged = merged.merge(nurse_assess_summary, on='patientunitstayid', how='left')
merged = merged.merge(apache_var_summary, on='patientunitstayid', how='left')
merged = merged.merge(vitals_cols_summary, on='patientunitstayid', how='left')
merged = merged.merge(medication_cols_sumary, on='patientunitstayid', how='left')
merged = merged.merge(custom_lab_summary, on='patientunitstayid', how='left')
merged = merged.merge(past_history_summary, on='patientunitstayid', how='left')
merged = merged.merge(infusion_drug_summary, on='patientunitstayid', how='left')
# ...existing code...
```

```python
# Convert to numeric and fill with median
for col in numeric_cols:
    if col in merged.columns:
        merged[col] = pd.to_numeric(merged[col], errors='coerce')
        merged[col].fillna(merged[col].median(), inplace=True)
```

```python
# Drop patient ID column if present
X = merged.drop(columns=['patientunitstayid'], errors='ignore')

# Keep only numeric columns for clustering
X = X.select_dtypes(include=['number'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Fill any remaining NaNs in X with column median
X = X.apply(lambda col: col.fillna(col.median()), axis=0)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# Let's start with 3 clusters (can tune later)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)

# Assign cluster labels
merged['cluster'] = kmeans.labels_
```

```python
# View how many patients per cluster
print(merged['cluster'].value_counts())

# Quick cluster means
cluster_summary = merged.groupby('cluster').mean(numeric_only=True)
print(cluster_summary)
```
